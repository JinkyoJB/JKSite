- **개요**: Random Forest는 여러 개의 의사결정 나무(Decision Trees)를 조합하여 작동하는 앙상블 학습 방법입니다.
- **작동 원리**: 각 나무는 부트스트랩 샘플(원래 데이터 셋에서 무작위로 추출된 샘플)을 기반으로 학습하고, 각 나무의 예측을 집계하여 최종 예측을 생성합니다.
- **장점**: 개별 트리의 과대적합 문제를 줄이고 일반화 성능을 향상시킵니다.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 예제 데이터를 로드하고 트레이닝/테스트 세트로 나눕니다:
iris = datasets.load_iris()
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 랜덤 포레스트 모델을 트레이닝합니다:
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# 테스트 데이터에 대한 예측을 수행합니다:
y_pred = clf.predict(X_test)

# 2D 그래프를 그리기 위한 함수
def plot_2d(X, y, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1)
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

# 3D 그래프를 그리기 위한 함수
def plot_3d(X, y, title):
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.Set1)
    ax.set_title(title)
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.set_zlabel('Feature 3')
    plt.show()

# 주요 특성 2개를 선택합니다 (2D 그래프용)
X_2d = X[:, :2]

# 주요 특성 3개를 선택합니다 (3D 그래프용)
X_3d = X[:, :3]

# 2D 그래프로 예측 결과를 표현합니다:
plot_2d(X_2d, y, 'Random Forest Predictions (2D)')

# 3D 그래프로 예측 결과를 표현합니다:
plot_3d(X_3d, y, 'Random Forest Predictions (3D)')
```
